### Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Plot Single Python List

```python
list = [3, 7, 8, 9, 12]
plt.plot(list)
plt.title('Single Python Plot')
plt.xlabel('Index')
plt.ylabel('Number')
plt.show()
```

### Plot Two Python Lists

```python
x = [1, 4, 7, 10, 14]
y = [2, 7, 8, 3, 16]
```

```python
plt.plot(x)
plt.title('Single Python Plot for X List')
plt.xlabel('Index')
plt.ylabel('Number')
plt.show()
```

```python
plt.plot(y)
plt.title('Single Python Plot for Y List')
plt.xlabel('Index')
plt.ylabel('Number')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Single Plot for X and Y')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

```python
fig,ax1 = plt.subplots()

ax1.plot(x, color = 'red')
ax1.set_title('Twin Axis Subplot of X and Y')
ax1.set_xlabel('Index')
ax1.set_ylabel('X', color = 'red')
ax1.tick_params(axis = 'y',
               labelcolor = 'red')

ax2=ax1.twinx()

ax2.plot(y, color = 'blue')
ax2.set_ylabel('Y', color = 'blue')
ax2.tick_params(axis = 'y',
               labelcolor = 'blue')

plt.show()
```

### 1D Array using Numpy

```python
x = np.linspace(0, 100, num = 100)
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2)
ax.set_title('Single Plot for X and X**2')
ax.set_xlabel('X')
ax.set_ylabel('X**2')
plt.show()
```

```python
fig,ax1 = plt.subplots()

ax1.plot(x, color = 'red')
ax1.set_title('Twin Axis Subplot of X and X**2')
ax1.set_xlabel('Index')
ax1.set_ylabel('X', color = 'red')
ax1.tick_params(axis = 'y',
               labelcolor = 'red')

ax2=ax1.twinx()

ax2.plot(x**2, color = 'blue')
ax2.set_ylabel('X**2', color = 'blue')
ax2.tick_params(axis = 'y',
               labelcolor = 'blue')

plt.show()
```

```python
plt.scatter(x, np.exp(x))
plt.title('Scatter Plot of X and Exponential of X')
plt.xlabel('X')
plt.ylabel('Exponential of X')
plt.show()
```

### Removing Symbols from Price

```python
car_sales['Price'] = car_sales['Price'].str.replace('[$, ,]', '', regex=True)
```

```python
car_sales.head()
```

```python
car_sales['Price'] = car_sales['Price'].astype(np.int64)
```

```python
car_sales.info()
```

### Add Total Sales Column in Car Sales DF

```python
car_sales['Total_Sales'] = car_sales['Price'].cumsum()
```

```python
car_sales.head()
```

```python
car_sales.info()
```

### Add Column Sale Date in Car Sales DF

```python
car_sales['Sale_Date'] = pd.date_range(start ='2-19-2024', 
                                       periods = len(car_sales))
```

```python
car_sales.head()
```

```python
car_sales.tail()
```

```python
car_sales.info()
```

### Price vs Total Sales Line Plot

```python
plt.plot(car_sales['Price'], car_sales['Total_Sales'])
plt.title('Price vs Total Sales Line Plot')
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.show()
```

### Odometer (KM) vs Price Scatter Plot

```python
plt.scatter(car_sales['Odometer (KM)'], 
           car_sales['Price'])
plt.title('Odometer (KM) vs Price Scatter Plot')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.show()
```

### Make vs Odometer (KM) Bar Plot

```python
plt.bar(car_sales['Make'], car_sales['Odometer (KM)'])
plt.title('Make vs Odometer (KM) Bar Plot')
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.show()
```

### Histogram of Make Column

```python
plt.hist(car_sales['Make'])
plt.xlabel("Make")
plt.ylabel("Frequency")
plt.title("Histogram Plot of Make")
plt.show()
```

### Subplot of Multiple Plots from Previous Steps

```python
fig=plt.figure(figsize=(12,12))

ax1=fig.add_subplot(221)
ax1.set_title("Price vs Total Sales Line Plot")
ax1.plot(car_sales['Price'],
        car_sales['Total_Sales'])

ax2=fig.add_subplot(222)
ax2.set_title("Odometer (KM) vs Price Scatter Plot")
ax2.scatter(car_sales['Odometer (KM)'],
        car_sales['Price'])

ax3=fig.add_subplot(223)
ax3.set_title("Make vs Odometer (KM) Bar Plot")
ax3.bar(car_sales['Make'],
        car_sales['Odometer (KM)'])

ax4=fig.add_subplot(224)
ax4.set_title("Histogram Plot of Make")
ax4.hist(car_sales['Make'])

plt.show()
```

### Color vs Total Sales Box Plot

```python
sns.set(rc = {'figure.figsize': (10, 9)})
sns.boxplot(x = 'Colour', y = "Total_Sales", data = car_sales)
plt.show()
```

### Total Sales and Odometer (KM) using Linear Regression

```python
sns.set(rc = {'figure.figsize': (15, 10)})
sns.lmplot(x = 'Total_Sales', y = 'Odometer (KM)', hue = 'Make', 
           ci = None, data = car_sales)
plt.show()
```

### Pair Plot for All Possible Combinations

```python
sns.set(rc = {'figure.figsize': (10, 9)})
sns.pairplot(car_sales[['Odometer (KM)', 'Doors', 'Price', 'Total_Sales']])
plt.show()
```

### Correlation between Total Sales, Price, and Odometer (KM)

```python
corr = car_sales[['Total_Sales', 'Price', 'Odometer (KM)']].corr()
print(corr)
```

```python
sns.set(rc = {'figure.figsize': (10, 9)})
sns.heatmap(corr, annot = True)
plt.show()
```
